```python
import pandas as pd
import numpy as np

 #Creating dummy arrays, lists, dictionary, tuples, etc...

arr1 = np.arange(12,21,2) #2d ndarray of even numbers
arr2 = np.arange(11,21,2) #2d ndarray of odd numbers
dicOfArray = {
    'a': arr1,
    'b':arr2
}
dicOfList = {
    'c': ['cat','copper','commor','commet', 'comedy'],
    'd': ['dog', 'dig','dome','dora', 'dosa']
}
li1 = [i for i in range(51,61,2)] #list of 5 odd numbers
li2 = [i for i in range(52,61,2)]

tup1 = ('one','two','three','four','five')
tup2 = ('first', 'second','third','fourth','five')

s1 = pd.Series(['onion','potato','catato', 'natato', 'jomi'])
s2 = pd.Series([i for i in range(71,81,2)])
 
dicOfSeries = {
    's1':s1,
    's2':s2
}
dicOfDic = {
    'dic1':dicOfArray,
    'dic2':dicOfList
}

dfOfList = pd.DataFrame(['apple','banana','cat','doberman','eclipse'])
dfOfNumbers = pd.DataFrame([i for i in range(41,51,2)])
dfOfArray = pd.DataFrame(arr1)
print(f'array:\n {arr1}\n {arr2}\ndictionaries of array, dic, list and series: \n{dicOfArray}\n {dicOfDic}\n {dicOfList}\n {dicOfSeries}\n {dfOfList} \n {dfOfArray} \n{dfOfNumbers}\n ')

#creating series from different data structures
seriesFromRange = pd.Series(range(5))
print('series from range :\n', seriesFromRange)

seriesFromArray = pd.Series(arr1)
print('series from array, :\n',seriesFromArray)

seriesFromSeries = pd.Series(seriesFromRange)
print('series from series: \n',seriesFromSeries)

seriesFromDic = pd.Series(dicOfArray)
print('series from dictionary :\n',seriesFromDic)

seriesFromTup = pd.Series(tup1)
print('series from tuple: \n',seriesFromTup)

#creating dictonary from various data structures.
dicFromArray = pd.DataFrame(arr2)
print('dictionary from array:\n',dicFromArray)

dicFromSeries = pd.DataFrame(seriesFromArray)
print('dictionary from Series:\n',dicFromSeries)

dicFromDic = pd.DataFrame(dicFromArray)
print('dictionary form dictionary: \n', dicFromDic)

dicFromTup = pd.DataFrame(tup1)
print('dictionary from tuple:\n',dicFromTup)

dicFromList = pd.DataFrame(li1)
print('dictionary from list:\n',dicFromList)

dicWithCustomColumn = pd.DataFrame(['karan','saran','maran', 'jaran'], columns=['name'], index=['first','second','third','fourth'])
print('dictionary with custom index and colum:\n',dicWithCustomColumn)

#USING INDEXING

print('getting first 2 values from series:\n', seriesFromRange[:2])
print('printing every second element: \n',seriesFromArray[::2])

arr3 = np.append(arr1,arr2)
print('appending two arrays:\n', arr3)

#using difference function
i1 = pd.Index([3,4,5,2])
i2 = pd.Index([4,5,33,22])
differencei = i1.difference(i2)
print('Difference func returns the items of the first list that are not in the second list:\n', differencei)

#using intersection function
intersectioni = i1.intersection(i2)
print('Intersection of i1 and i2:\n',intersectioni)

#union function
unioni = i1.union(i2)
print('union of i1 and i2 (it returns sorted result)',unioni)

isinreally = s1.isin([2,23,23,43,3])
print('isin function prints true if element in arr1 is in []/arr2.\n',isinreally)

remainingarray = np.delete(arr1,1)
print('this deletes elements from the given array. [original array] [modified array]\n',arr1, remainingarray)

dropf = dfOfArray.drop(0,axis=0)
print('this works on dataframe [original dataframe] [new dataframe]\n',dfOfArray, dropf)

#insert function works on both array and dataframe
dfOfArray.insert(1,'cats',['ginger','loki','karu','baru','moki'])
print('new column inserted in dfofcolumn.\n',dfOfArray )

print('is_monotonic is depreciated and is replaced by is_monotonic_decreasing/increasing\n',i1.is_monotonic_decreasing, i1.is_monotonic_increasing)

print('is_unique check is each element in a series or dataframe is unique or not\n', s1.is_unique)

print('unique function returns an array of unique element from a series or dataframe\n', s2.unique() )

# reindexing
print(dfOfList)
cdfol = dfOfList.reindex([4,3,0,1,2,0])
print(cdfol)

#selecting and filtering
dfOfArray.columns=['age','cats']
print(dfOfArray)
print(dfOfArray[dfOfArray['age']>15]) #this will return the rows where age > 15

#using lable based indexing and index based indexing
dfOfArray.index = ['east','west','north','south','center']
print(dfOfArray.loc['east']) #outputs column and row at row lable east

print(dfOfArray.iloc[-1]) #this outputs last cell

#Arithmetic and Data Alignment (add, radd, sub, rsub, div, rdiv, floordiv, rfloordiv, mul, rmul, pow, rpow.)
print(s1+" cat")
print(arr1, 'updated', arr1+0.5)

#'r'opr. means reverse, what it basically does is. obj1.add(obj2)->obj1+obj2, obj1.r[opr](obj2) -> obj2+obj1
arrayadd  = arr1.__add__(arr2)
print(arr1,arr2,arrayadd)

arrayminus = arr1.__sub__(arr2)
arrayrevminus = arr1.__rsub__(arr2)
print(arrayminus,arrayrevminus)



```
